```python
#f
import random

# Define the set A
A = [0, 1, 2, 5, 6, 9]

# Randomly select a, b, c, d from A
a, b, c, d = random.sample(A, 4)

# Print the selected values
print("Selected values:")
print(f"a = {a}, b = {b}, c = {c}, d = {d}")


# Create the digraph
D_f = DiGraph([(a, a), (a, b), (a, c), (a, d), (d, c), (c, d), (b, c), (b, d), (d, d), (b, b)], loops=True)

# Show the digraph with customized options
D_f.show(figsize=[10, 10], vertex_size=300, vertex_labels=True)
```
